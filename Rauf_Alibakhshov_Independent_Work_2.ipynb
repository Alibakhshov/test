## Task 1: Linear Approximation
<p>Implement a function that takes a function and a value, and returns a linear approximation of that function at that value.</p>

```python
from sympy import*
import matplotlib.pyplot as plt
%matplotlib inline
x = symbols('x')

# Taking inputs
func = input("Enter the function: ")
a = float(input("Enter a: "))
value = float(input("Enter a number for evaluation: "))

# f of a
f = lambdify (x, func)
f_of_a = f(a)

# f derivative of a
expr_diff = Derivative(func, x).doit()
f_der = lambdify (x, expr_diff)
f_der_of_x = f_der(a)

# Final equation 
linear_expr = f_of_a + (f_der_of_x * (x-a))
linear_func = lambdify(x, linear_expr)
result = linear_func(value)

# Printing the result
print(result)

# Drawing the graphs
a = [i for i in range(0,30)]
y = [f(i) for i in a]
y_new = [linear_func(i) for i in a]
plt.figure(figsize = (20,10))
plt.plot(a, y, label = func)
plt.plot(a, y_new, label = linear_expr)

plt.show()



```

## Task 2: Newton's Method
<p>Implement a function that takes a differentiable function and returns any root of that function.</p>

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy.abc import x
from sympy import *
x = sym.symbols('x')

# inputing a function
func = input("Enter a function: ")

# inputing a value for x initial
x_initial = int(input("Enter x initial: "))

#converting to a function
f = sym.lambdify(x, func)

#plotting a graph of function
x_axis = np.linspace(-1, 5, 20)
plt.figure(figsize = (20,10))
plt.plot(x_axis, f(x_axis), linewidth = 3)

#taking derivative
expr_diff = Derivative(func, x).doit()
while abs(f(x_initial))>=0.00001:
    x1 = x_initial - f(x_initial)/expr_diff.subs(x, x_initial)
    plt.plot((x1, x_initial),(0, f(x_initial)), 'r--')
    x_initial = x1
plt.xlim(-1,5)
plt.ylim(0,20) 


```
